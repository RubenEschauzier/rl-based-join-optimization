```python
import os
import json
import numpy as np
import matplotlib.pyplot as plt

from main import ROOT_DIR
```

```python
def read_pretrain_results(location):
    with open(os.path.join(ROOT_DIR, location), 'rb') as f:
        return json.load(f)
path_no_literal = read_pretrain_results("data/pretrain_data/generated_queries/path_without_literal.json")
path_yes_literal = read_pretrain_results("data/pretrain_data/generated_queries/path_with_literal.json")

cardinalities_no_literal = np.array([query['y'] for query in path_no_literal if query and query['y'] < 2e5])
cardinalities_yes_literal = np.array([query['y'] for query in path_yes_literal if query and query['y'] < 1000])
```

```python
def plot_distribution(data):
    plt.hist(data, bins=50, color='skyblue', edgecolor='black')
    plt.title('Cardinality distribution')
    plt.xlabel('Cardinality')
    plt.ylabel('Frequency')
    plt.show()
plot_distribution(cardinalities_no_literal)
plot_distribution(cardinalities_yes_literal)
```
